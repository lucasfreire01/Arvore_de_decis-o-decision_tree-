```python
import pandas as pd
dataset = pd.read_csv('/content/credit_data.csv')
```

```python
dataset
```

```python
dataset.dropna(inplace=True)
dataset.shape
```

```python
x = dataset.iloc[:, 1:4].values
```

```python
x
```

```python
y = dataset.iloc[:, 4].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
from sklearn import naive_bayes
resultados_naive = []
resultados_logistica = []
resultados_random = []
for i in range(30):
  x_treinamento, x_teste, y_treinamento, y_teste = train_test_split(x,y, test_size=0.2,
                                                                    stratify=y, random_state=1)
  naive_bayes = GaussianNB()
  naive_bayes.fit(x_treinamento, y_treinamento)
  resultados_naive.append(accuracy_score(y_teste, naive_bayes.predict(x_teste)))

  logistica = LogisticRegression()
  logistica.fit(x_treinamento, y_treinamento)
  resultados_logistica.append(accuracy_score(y_teste, logistica.predict(x_teste)))

  random_florest = RandomForestClassifier()
  random_florest.fit(x_treinamento, y_treinamento)
  resultados_random.append(accuracy_score(y_teste, random_florest.predict(x_teste)))
```
